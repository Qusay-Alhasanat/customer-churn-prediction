# Final Conclusion & Business Insights


The objective of this project was to predict customer churn using historical customer data and to identify patterns that help reduce customer loss.


In this project, we:
- Cleaned and prepared the data
- Performed exploratory data analysis (EDA)
- Engineered relevant features
- Trained and evaluated machine learning models


Key insights from the analysis:
- Contract type and tenure have a strong impact on churn
- Customers with short tenure are more likely to churn
- Class imbalance significantly affects model performance


The final model selected was Logistic Regression with class balancing.

This model prioritizes detecting churned customers, which is critical for:
- Retention campaigns
- Personalized offers
- Reducing long-term revenue loss


Possible future improvements:
- Try advanced models (Random Forest, XGBoost)
- Perform hyperparameter tuning
- Deploy the model as an API or dashboard
